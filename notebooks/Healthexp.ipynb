```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset("healthexp")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grfico de linha para visualizar a tendncia de gastos com sade ao longo do tempo para cada pas
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='Year', y='Spending_USD', hue='Country', marker='o')
plt.title('Tendncia de Gastos com Sade ao Longo do Tempo por Pas')
plt.xlabel('Ano')
plt.ylabel('Gastos com Sade (USD)')
plt.legend(title='Pas', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()

# Grfico de linha para visualizar a tendncia de expectativa de vida ao longo do tempo para cada pas
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='Year', y='Life_Expectancy', hue='Country', marker='o')
plt.title('Tendncia de Expectativa de Vida ao Longo do Tempo por Pas')
plt.xlabel('Ano')
plt.ylabel('Expectativa de Vida')
plt.legend(title='Pas', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estatsticas resumidas dos gastos com sade por pas
summary_stats = df.groupby('Country')['Spending_USD'].describe()

# Grfico de barras para comparar os gastos com sade entre os pases
plt.figure(figsize=(12, 8))
sns.barplot(data=summary_stats, x=summary_stats.index, y='mean', palette='viridis')
plt.title('Comparao de Gastos com Sade entre Pases (Mdia)')
plt.xlabel('Pas')
plt.ylabel('Gastos com Sade (Mdia)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Grfico de boxplot para visualizar a distribuio dos gastos com sade por pas
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Country', y='Spending_USD', palette='viridis')
plt.title('Distribuio dos Gastos com Sade por Pas')
plt.xlabel('Pas')
plt.ylabel('Gastos com Sade (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definindo o tamanho da figura
plt.figure(figsize=(14, 6))

# Grfico de boxplot para os gastos com sade
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='Country', y='Spending_USD', color='skyblue')
plt.title('Boxplot dos Gastos com Sade por Pas')
plt.xlabel('Pas')
plt.ylabel('Gastos com Sade (USD)')

# Identificando e destacando outliers nos gastos com sade
outliers_spending = df[df['Spending_USD'] > df['Spending_USD'].quantile(0.95)]
plt.scatter(outliers_spending['Country'], outliers_spending['Spending_USD'], marker='o', color='red', label='Outliers')
plt.legend()

# Grfico de boxplot para a expectativa de vida
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='Country', y='Life_Expectancy', color='lightgreen')
plt.title('Boxplot da Expectativa de Vida por Pas')
plt.xlabel('Pas')
plt.ylabel('Expectativa de Vida')

# Identificando e destacando outliers na expectativa de vida
outliers_life = df[df['Life_Expectancy'] < df['Life_Expectancy'].quantile(0.05)]
plt.scatter(outliers_life['Country'], outliers_life['Life_Expectancy'], marker='o', color='red', label='Outliers')
plt.legend()

plt.xticks(rotation=45)  # Rotaciona os rtulos do eixo x para melhorar a legibilidade

plt.tight_layout()
plt.show()
```
