```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset("healthexp")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gráfico de linha para visualizar a tendência de gastos com saúde ao longo do tempo para cada país
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='Year', y='Spending_USD', hue='Country', marker='o')
plt.title('Tendência de Gastos com Saúde ao Longo do Tempo por País')
plt.xlabel('Ano')
plt.ylabel('Gastos com Saúde (USD)')
plt.legend(title='País', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()

# Gráfico de linha para visualizar a tendência de expectativa de vida ao longo do tempo para cada país
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='Year', y='Life_Expectancy', hue='Country', marker='o')
plt.title('Tendência de Expectativa de Vida ao Longo do Tempo por País')
plt.xlabel('Ano')
plt.ylabel('Expectativa de Vida')
plt.legend(title='País', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estatísticas resumidas dos gastos com saúde por país
summary_stats = df.groupby('Country')['Spending_USD'].describe()

# Gráfico de barras para comparar os gastos com saúde entre os países
plt.figure(figsize=(12, 8))
sns.barplot(data=summary_stats, x=summary_stats.index, y='mean', palette='viridis')
plt.title('Comparação de Gastos com Saúde entre Países (Média)')
plt.xlabel('País')
plt.ylabel('Gastos com Saúde (Média)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Gráfico de boxplot para visualizar a distribuição dos gastos com saúde por país
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Country', y='Spending_USD', palette='viridis')
plt.title('Distribuição dos Gastos com Saúde por País')
plt.xlabel('País')
plt.ylabel('Gastos com Saúde (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definindo o tamanho da figura
plt.figure(figsize=(14, 6))

# Gráfico de boxplot para os gastos com saúde
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='Country', y='Spending_USD', color='skyblue')
plt.title('Boxplot dos Gastos com Saúde por País')
plt.xlabel('País')
plt.ylabel('Gastos com Saúde (USD)')

# Identificando e destacando outliers nos gastos com saúde
outliers_spending = df[df['Spending_USD'] > df['Spending_USD'].quantile(0.95)]
plt.scatter(outliers_spending['Country'], outliers_spending['Spending_USD'], marker='o', color='red', label='Outliers')
plt.legend()

# Gráfico de boxplot para a expectativa de vida
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='Country', y='Life_Expectancy', color='lightgreen')
plt.title('Boxplot da Expectativa de Vida por País')
plt.xlabel('País')
plt.ylabel('Expectativa de Vida')

# Identificando e destacando outliers na expectativa de vida
outliers_life = df[df['Life_Expectancy'] < df['Life_Expectancy'].quantile(0.05)]
plt.scatter(outliers_life['Country'], outliers_life['Life_Expectancy'], marker='o', color='red', label='Outliers')
plt.legend()

plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x para melhorar a legibilidade

plt.tight_layout()
plt.show()
```
